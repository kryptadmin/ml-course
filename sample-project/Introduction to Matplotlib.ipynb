## Matplotlib example workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple plot", xlabel="X-axis", ylabel="Y-axis")

# 5. Save and show (you save whole figure)
fig.savefig("./images/sample-plot.png")
```

# Two options for subplots

```python
# subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.random(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Plot to each different axes
ax[0,0].plot(x, x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000))
```

### Which one should you use? (pyplot vs matplotlib OO method?)

* When plotting something quickly, okay to use the pyplot method
* When plotting something more advanced use  OO method

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
over_50
```

```python
# Pyplot method
over_50.plot(kind="scatter", x="age", y="chol", c="target");
```

```python
# OO method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax);
# ax.set_xlim([45, 100])
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])

# Customize the plot
ax.set(title="Heart disease and cholestrol levels", xlabel="age", ylabel="Cholestrol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle="--")
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(figsize=(10, 10), nrows=2, ncols=1, sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])

# Customize ax0
ax0.set(title="Heart disease and cholestrol", ylabel="cholestrol")

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax0.axhline(y=over_50["chol"].mean(), linestyle="--")

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"])

# Customize ax1
ax1.set(title="Heart disease and Talach", ylabel="talach")

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--")

# Add a title to the figure
fig.suptitle("Heart disease analysis", fontsize=16, fontweight="bold")
```

```python
# Set the style
plt.style.use("seaborn-v0_8-whitegrid")

# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"], cmap="gist_gray")

# Customize the plot
ax.set(title="Heart disease and cholestrol levels", xlabel="age", ylabel="Cholestrol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle="--")
```

```python
# Customizing x and y axis limitations
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(figsize=(10, 10), nrows=2, ncols=1, sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"], cmap="winter")

# Customize ax0
ax0.set(title="Heart disease and cholestrol", ylabel="cholestrol")
# Change the x-axis
ax0.set_xlim([50,80])

# Change the ylimit
ax0.set_ylim([60, 500])

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax0.axhline(y=over_50["chol"].mean(), linestyle="--")

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"], cmap="winter")

# Customize ax1
ax1.set(title="Heart disease and Talach", ylabel="talach")
# Change the x-axis
ax1.set_xlim([50, 80])

# Change the y lim
ax1.set_ylim([60, 200])

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--")

# Add a title to the figure
fig.suptitle("Heart disease analysis", fontsize=16, fontweight="bold")
```
